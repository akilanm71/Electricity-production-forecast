```python
import seaborn as sb
```

```python
plt.scatter(df["DayTon_MV"],df["DayTon_MV(t-1)"], color ="orange")
plt.title("Covarriance Plot of Hourly shift of Electricity Production")
plt.show()
```

## The points form a diagonal cloud from bottom left to top right
## Indicates strong lag 1 autocorrelation

```python
from pandas.plotting import autocorrelation_plot
auto_co = autocorrelation_plot(df["DayTon_MV"])
```

## Grey Horizontal band indicates 95% Confidence interval
## Data is Highly Correlated, Past values strongly influence future values, even many hours later

```python
## Function to remove trend by taking shift 1
def sdi():
    x = df["DayTon_MV"].values
    diff = []
    for i in range(1, len(x)):
        y =  x[i]-x[i-1]
        diff.append(y)
    return diff
```

```python
dfdt = sdi()
```

```python
dfdt = pd.Series(dfdt)
```

```python
len(dfdt)
```

```python
dfdt.plot(color ='#FF5734')
plt.show()
```

# Intrepretation of Plot
## Mean around Zero (Trend has been removed)
## No sytamatic upward and downward shift confirms no Trend and seasonality
## Confirms that stationary can be achieved
## Variance is roughly constant some have large spikes (mild heteroskedasticity)

```python
plt.figure()
fig, axis = plt.subplots(1,2)
autocorrelation_plot(dfdt, ax =axis[0],color ="red")
autocorrelation_plot(dfdt[100:200], ax= axis[1])
axis[0].set_title("Full Series")
axis[1].set_title("Subset (100–200)")
```

## ACF now decays quickly

```python
def evaluate_arima(test_data, arima_order):
    history = list(train_q2)
    predictions = []

    for t in range(len(test_data)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        print(f"Models step {test_data[t]} is {model_fit.aic}")
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)

        obs = test_data.iloc[t]
        history.append(obs)

    rmse = sqrt(mean_squared_error(test_data, predictions))
    return rmse
```

```python
def arima_grid():
    p_li = [0, 1, 2]
    d_li = [0, 1]
    q_li = [0, 1, 2]

    best_score, best_cfg = float("inf"), None

    for p in p_li:
        for d in d_li:
            for q in q_li:
                order = (p, d, q)
                try:
                    rmse = evaluate_arima(test_q2, order)

                    if rmse < best_score:
                        best_score, best_cfg = rmse, order

                    print(f"ARIMA{order} RMSE={rmse:.3f}")

                except Exception as e:
                    print(f"ARIMA{order} failed: {e}")

    print(f"\nBest ARIMA{best_cfg} RMSE={best_score:.3f}")




```

```python
grid_par = arima_grid()
```

```python
import tensorflow as tf
```

```python
total_size = len(xtt1)
train_size = int(0.7*total_size)
val_size = int(0.2*total_size)
test_size = total_size - (train_size+val_size)
```

```python
train_size ,val_size , test_size
```

```python
train = xtt1.iloc[:train_size]
val = xtt1.iloc[train_size:val_size+train_size]
test =  xtt1.iloc[val_size+train_size:]
```

```python
def split_window(series, window):
    X = []
    y = []
    for i in range(len(series) - window):
        X.append(series.iloc[i:i+window])
        y.append(series.iloc[i+window])
    return tf.stack(X), tf.stack(y)


```

```python
train
```

```python
val
```

```python
test
```

```python
window = 24
x_train, y_train = split_window(train, window)
```

```python
x_val, y_val = split_window(val, window)
x_test, y_test = split_window(test, window)
```

```python
x_train = tf.expand_dims(x_train, axis=-1)
x_val   = tf.expand_dims(x_val, axis=-1)
x_test  = tf.expand_dims(x_test, axis=-1)

```
